# Création Dataset avec colonnes handicap et âge

```python
data = np.ones(shape=(100000, 1))
df = pd.DataFrame(data, columns=['x'])
# Création colonne handicap avec des NaN
df['handicap'] = pd.NA

# Liste d'handicaps
liste_handicap = ['Sensoriel', 'Physique', 'Cognitif', 'Fonctionnel']

# Probabilités pour chaque handicap : ['Sensoriel', 'Physique', 'Cognitif', 'Fonctionnel']
probalite_handicap = [0.17, 0.27, 0.13, 0.43]

# Création colonne handicap 
df['handicap'] = df['handicap'].apply(lambda x: np.random.choice(liste_handicap, p=probalite_handicap))

# Verify
df['handicap'].value_counts()

# Listes probabilités de catégories d'ages par handicap : ['Sensoriel', 'Physique', 'Cognitif', 'Fonctionnel'] 
liste_probabilite_age = [[0.08, 0.31, 0.61], [0.04, 0.29, 0.67], [0.23, 0.32, 0.45], [0.10, 0.57,0.33]]

# Liste des catégories d'âges
categorie_age = ['5-25ans' ,'25-59ans', '60+']

# Création colonne age avec des NaN
df['age'] = pd.NA

# Remplacement des NaN de la colonne 'age' pour chaque handicap avec les probabilités associées à l'handicap 

for x, y in zip(liste_handicap, liste_probabilite_age):
    df.loc[df['handicap'] == x, 'age']  = df.loc[df['handicap'] == x, 'age'].apply(lambda x: np.random.choice(categorie_age, p=y))
```
